```python
import json
import os
```

```python
lst = []
for name in os.listdir():
    if not name.endswith('.json'):
        continue
    with open(name, 'r') as json_file:
        lst.append(json.load(json_file))
```

–ß—Ç–æ–±—ã "–ø—Ä–µ–≤—Ä–∞—Ç–∏—Ç—å" json-—Ñ–∞–π–ª/json-–æ—Ç–≤–µ—Ç –≤ —Å–ª–æ–≤–∞—Ä—å, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è —Ñ—É–Ω–∫—Ü–∏–µ–π load/loads

–í –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ –±—ã–ª –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω —Ü–∏–∫–ª –ø–æ –≤—Å–µ–º —Ñ–∞–π–ª–∞–º –≤ –ø–∞–ø–∫–µ, –∏, –µ—Å–ª–∏ –æ–Ω–∏ –∏–º–µ—é—Ç —Ä–∞—Å—à–∏—Ä–µ–Ω–∏–µ .json, —Ç–æ –∑–∞–≥—Ä—É–∂–∞–µ—Ç—Å—è –¥–∞–Ω–Ω—ã–π —Ñ–∞–π–ª –≤ —Å–ø–∏—Å–æ–∫

```python
for i in lst:
    try:
        print(f"ID: {i['graphql']['user']['id']}")
        print(f"Full name:  {i['graphql']['user']['full_name']}")
        print(f"Status:  {i['graphql']['user']['biography']}")
        print(F"Is business account: {i['graphql']['user']['is_business_account']}")
        print(f"Avatar: {i['graphql']['user']['profile_pic_url_hd']}")
        print()
        print("POSTS")
        for j in i['graphql']['user']['edge_owner_to_timeline_media']['edges']:
            print(j['node']['id'])
            print(j['node']['display_url'])
            try:
                print(j['node']['edge_media_to_caption']['edges'][0]['node']['text'])
            except IndexError:
                pass
            print('*'*30)
        print('-'*30)
    except TypeError:
        pass
```

–¢–µ–ø–µ—Ä—å –∏–∑ —ç—Ç–∏—Ö –¥–∞–Ω–Ω—ã—Ö —Å–¥–µ–ª–∞–µ–º –æ–±—Ä–∞—Ç–Ω–æ json –±–µ–∑ —Å–ª—É–∂–µ–±–Ω–æ–π –∏ –ª–∏—à–Ω–µ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏

```python
lst1 = []
for i in lst:
    try:
        dicti = {}
        dicti['inst_id'] = i['graphql']['user']['id']
        dicti['full_name'] = i['graphql']['user']['full_name']
        dicti['status'] = i['graphql']['user']['biography']
        dicti['business_boolean'] = i['graphql']['user']['is_business_account']
        dicti['avatar_src'] = i['graphql']['user']['profile_pic_url_hd']
        dicti['posts'] = []
        for j in i['graphql']['user']['edge_owner_to_timeline_media']['edges']:
            try:
                dicti['posts'].append({
                    'post_id': j['node']['id'],
                    'photo_src': j['node']['display_url'],
                    'text': j['node']['edge_media_to_caption']['edges'][0]['node']['text'],
                })
            except IndexError:
                dicti['posts'].append({
                    'post_id': j['node']['id'],
                    'photo_src': j['node']['display_url'],
                })
        lst1.append(dicti)
    except TypeError:
        pass
```

```python
with open('inst.json', 'w') as f:
    json.dump(lst1, f)
```

```python
with open('inst.json', 'r') as f:
    print(json.load(f))
```
